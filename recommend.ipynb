```python
import pandas as pd
import numpy as  np
```

```python
movies = pd.read_csv("C:/Users/PAPA KA PARA/Downloads/top10K-TMDB-movies.csv.csv")
movies.head()
```

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'genre',  'overview']]
```

```python
movies.head()
```

```python
movies['tags'] = movies['genre'] +  movies['overview']
```

```python
movies.head()
```

```python
new_data = movies.drop(columns = ['genre' ,'overview'])
```

```python
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 10000 ,stop_words ='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new_data[new_data['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])) , reverse = True , key = lambda vector :vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Schindler's List")
```
